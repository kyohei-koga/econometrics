```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
%matplotlib inline
```

```python
df = pd.read_excel('./5-5.xls')
```

```python
df.columns = df.iloc[2, :]
df = df.iloc[3:, :]
df = df.astype({'為替レート（EXR）': float, '消費者物価指数（CPI）': float, '賃金率（WAGE）': float})
```

```python
# 線形モデルの推定
y = df.iloc[:, 2]
x = df.loc[:, ['為替レート（EXR）', '賃金率（WAGE）']]
x_c =sm.add_constant(x)
model = sm.OLS(y ,x_c)
result = model.fit()
result.summary()
```

```python
# 対数線形モデルの推定
y = np.log(df.iloc[:, 2])
x = np.log(df.loc[:, ['為替レート（EXR）', '賃金率（WAGE）']])
x_c =sm.add_constant(x)
model = sm.OLS(y ,x_c)
result = model.fit()
result.summary()
```
